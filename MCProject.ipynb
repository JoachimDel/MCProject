## 1 - Simulation de trajectoires en dimension 2

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def dimension_2_random_walk(X):
    U = np.random.random()
    if U > 0.5:
        if U > 0.75:
            X[0] += 1
        else:
            X[0] -= 1
    else:
        if U > 0.25:
            X[1] += 1
        else:
            X[1] -= 1
```

```python
n = 30000
X = [0,0]
trajectoire = []
trajectoire.append(tuple(X))
longueurs = []
nbr_sections = 0
longueur = 0

while nbr_sections < n:    
    dimension_2_random_walk(X)

    if tuple(X) in trajectoire:
        longueurs.append(longueur)
        nbr_sections += 1
        longueur = 0
    else:
        longueur += 1
    
    trajectoire.append(tuple(X))
```

```python
x = list()
y = list()

for i in range(len(trajectoire)):
    x.append(trajectoire[i][0])
    y.append(trajectoire[i][1])

fig, (ax1,ax2) = plt.subplots(1,2,figsize = (18,5))
ax1.plot(x,y)
ax2.hist(longueurs,bins=range(20))
plt.show()
```
